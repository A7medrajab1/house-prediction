 HANDLE MISSING VALUES

```python
df['total_bedrooms'].fillna(df['total_bedrooms'].mean(),inplace=True)
```

Mapping

```python
df['ocean_proximity'].value_counts()
```

```python
mapping={'<1H OCEAN':1,'INLAND':2,'NEAR OCEAN':3,'NEAR BAY':4,'ISLAND':5}
df['ocean_proximity']=df['ocean_proximity'].map(mapping)
```

BUILD A MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

### x: Contains all the independent variables (features) that the model will use to make predictions
### y: Contains the dependent variable (target) that model predicts it.

```python
x=df.drop(columns=['median_house_value']).values
y=df['median_house_value'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
KNN=KNeighborsRegressor()
KNN.fit(x_train,y_train)
```

RMSE BY KNN

```python
y_pred_knn=KNN.predict(x_test)
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred_knn))
mse = mean_squared_error(y_test,y_pred_knn)
print(mse)
print(rmse)
```

 RMSE BY LR

```python
y_pred_lr=LR.predict(x_test)
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred_lr))
mse = mean_squared_error(y_test,y_pred_lr)
print(mse)
print(rmse)
```
